Birank

```python
import birankpy
bn = birankpy.BipartiteNetwork()

def gen_birank(weight=None, normalizer=None):
    bn.set_edgelist(df, top_col='business_name', bottom_col='user_name', weight_col=weight)
    top_birank_df, bottom_birank_df = bn.generate_birank(normalizer = normalizer) if normalizer != None else bn.generate_birank() 
    return top_birank_df.sort_values(by='business_name_birank', ascending=False).reset_index(drop=True)
```

```python
output = gen_birank('weight')
```

```python
output['rank'] = output.index + 1

N = len(output)
def pr(s):
    pr_val = int((N - s)*100 / N)
    return pr_val

output['pr'] = output['rank'].apply(pr)
output
```
