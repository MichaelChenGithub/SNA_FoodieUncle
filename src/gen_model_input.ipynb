```python
import pandas as pd
```

```python
review = pd.read_csv('/Users/chenchw/Downloads/2018_lasvegas_bar_elite_review.csv',encoding = "ISO-8859-1")
review_df = pd.DataFrame(review)
review_df
```

```python

business = pd.read_json('/Users/chenchw/Downloads/business.json',lines = True)
user = pd.read_json('/Users/chenchw/Downloads/user.json',lines = True)
```

```python
res = pd.merge(review,business, on='business_id',how='left')
result = pd.merge(res,user,on='user_id',how='left')
```

```python
df = pd.DataFrame(result,columns=['name_x','name_y','stars_x', 'useful_x', 'funny_x',
       'cool_x'])
df.columns=['business_name','user_name','stars','useful','funny','cool']
df['weight'] = df['useful']+df['funny']+df['cool']
df_result = pd.DataFrame(df,columns=['business_name','user_name','stars','weight'])
```

```python
df_result.sort_values(by=['user_name','business_name'])
print(df_result)
```

```python
row1 = None
row0 = None
row10 = None
for row in df_result.columns:
    if row0 == row[0] and row1 == row[1]:
        row[-1]+=row10       
    else:
        row10 = row[-1]
        row0 = row[0]
        row1 = row[1]
```

```python
df_result = df_result.drop_duplicates(
    subset=['business_name','user_name'], keep="last")
```
